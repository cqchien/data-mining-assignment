```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('./job-market.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()].shape[0]
```

```python
df.drop_duplicates(inplace= True)
```

```python
df[df.duplicated()].shape[0]
```

```python
df.isnull().sum()
```

```python
df.dropna(how='all', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Location']
```

```python
df['NumberOfJob'] = 1
```

```python
df
```

```python
locationTable = pd.pivot_table(data=df, index=['Location'], aggfunc={'NumberOfJob': np.sum}).sort_values(by='NumberOfJob',ascending=False)

locationTable
```

```python
locationTableReset=locationTable.reset_index()
```

```python
locationTableReset
```

```python
sns.set(rc={'figure.figsize':(12,25)})

sns.barplot(x="NumberOfJob", y="Location",data=locationTableReset)
```
