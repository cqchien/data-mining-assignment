```python
pattern = re.compile('<.*?>')

def cleanHtml(html):
  cleanText =  re.sub(pattern, '', html)
  return cleanText
```

```python
for i in df.index:
  text = df['FullDescription'][i]
  df['FullDescription'][i] = cleanHtml(str(text))
```
